```
#solving the environment with ckiptagger
from ckiptagger import data_utils, construct_dictionary, WS, POS, NER
data_utils.download_data_gdown("./") # gdrive-ckip
ws = WS("./data")
```

```
import pandas as pd
import re
```

```
questions=pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/questions.csv')
```

```
#using the tmp buffer to store the processed data
tmp_corpus=[]
for i in range(len(questions['content'])):
    tmp=questions['content'][i]
    #remove \n to avoid line changing
    tmp=tmp.replace('\n', '')
    questions['content'][i]=re.sub("[A-Za-z0-9\!\\%[\]\,\。\，\.\？]", "", tmp)
    tmp_corpus.append(questions['content'][i])
```

```
#using ckiptagge to cut words in chinese
tmp_corpus=ws(tmp_corpus)
cut_corpus=[]

for i in tmp_corpus:
  cut_corpus.append(' '.join(i))

#alter the content after cutting the documents
for k in range(len(questions['content'])):
  questions['content'][k]=cut_corpus[k]



```

```
#save the processed data
questions.to_csv('/content/gdrive/My Drive/Colab Notebooks/cutquestions.csv',encoding='utf-8',index=False)
```
